```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
plt.style.use("seaborn-v0_8")
```

```python
X, y = make_blobs(n_samples = 2000, n_features = 2, centers = 3, cluster_std = 3, random_state = 42)
n = X.shape[1]
m = X.shape[0]
```

```python
X_test = np.array([-10, 2])


label_added = {0: False, 1: False, 2: False}

for i in range(m):
    if y[i] == 0:
        if not label_added[0]:
            plt.scatter(X[i, 0], X[i, 1], color = "red", label = "Class 0")
            label_added[0] = True

        else:
            plt.scatter(X[i, 0], X[i, 1], color = "red")

    elif y[i] == 1:
        if not label_added[1]:
            plt.scatter(X[i, 0], X[i, 1], color = "green", label = "Class 1")
            label_added[1] = True

        else:
            plt.scatter(X[i, 0], X[i, 1], color = "green")

    else:
        if not label_added[2]:
            plt.scatter(X[i, 0], X[i, 1], color = "blue", label = "Class 2")
            label_added[2] = True

        else:
            plt.scatter(X[i, 0], X[i, 1], color = "blue")

plt.scatter(X_test[0], X_test[1], color = "orange", marker = "*", label = "Test Point")
plt.legend()
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
# Euclidean Distance Formula.

def dist(p, q):
    return np.sqrt(np.sum((p - q) ** 2))
```

```python
# KNN Algorithm For Classification.

def KNN_classifcation(X, y, X_test, k = 5):
    m = X.shape[0]
    distances = []

    for i in range(m):
        d = dist(X[i], X_test)
        distances.append((d, y[i]))

    distances = sorted(distances)
    distances = np.array(distances)
    labels = distances[:k, 1]
    labels, counts = np.unique(labels, return_counts = True)
    index = counts.argmax()
    ans = labels[index]

    return int(ans)
```

```python
# KNN Algorithm For Regression.

def KNN_regression(X, y, X_test, k = 5):
    m = X.shape[0]
    distances = []

    for i in range(m):
        d = dist(X[i], X_test)
        distances.append((d, y[i]))

    distances = sorted(distances)
    distances = np.array(distances)
    labels = distances[:k, 1]
    ans = np.mean(labels)

    return ans
```

```python
KNN_classifcation(X, y, X_test)
```
