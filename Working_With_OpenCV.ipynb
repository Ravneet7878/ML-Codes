```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read The Image.

img = cv2.imread("flower image.jpg")
print(img.shape)
```

```python
# Display The Image.

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Converting BGR Pixel Format To RGB Pixel Format.

new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
plt.axis("off")
plt.show()
```
