```python
import numpy as np
import pandas as pd
```

```python
golf = pd.read_csv("golf.csv")
```

```python
golf
```

```python
# Calculating Prior Probability.

def prior_prob(golf, label):
    total_examples = golf.shape[0]
    class_examples = (golf["Play"] == label).sum()

    return class_examples / total_examples
```

```python
PRIOR = {
    "yes" : prior_prob(golf, "yes"),
    "no" : prior_prob(golf, "no")
}

print(PRIOR)
```

```python
# Calculating Conditional Probability.

def cond_prob(golf, feature, feature_value, label):
    filtered_data = golf[golf["Play"] == label]
    numerator = (filtered_data[feature] == feature_value).sum()
    denominator = (golf["Play"] == label).sum()

    return numerator / denominator
```

```python
features = list(golf.columns)[:-1]
COND_PROB = {}

for label in golf["Play"].unique():
    COND_PROB[label] = {}

    for feature in features:
        COND_PROB[label][feature] = {}

        for feature_value in golf[feature].unique():
            prob = cond_prob(golf, feature, feature_value, label)
            COND_PROB[label][feature][feature_value] = prob
            print(label, feature, feature_value, prob)

    print()
```

```python
# Prediction.

X_test = ["sunny", "hot", "normal", False]

denominator
numerator_yes
numerator_no

for label in golf["Play"].unique():
    prior = PRIOR[label]
    likelihood = 1.0

    for i in range(len(X_test)):
        feature = features[i]
        feature_value = X_test[i]

        likelihood *= COND_PROB[label][feature][feature_value]

    if label == "yes":
       numerator_yes = (likelihood * prior)

    else:
       numerator_no = (likelihood * prior)

    denominator += (likelihood * prior)

print("yes :", numerator_yes / denominator)
print("no :", numerator_no / denominator)
```
