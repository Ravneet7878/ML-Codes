```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("face image.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```

Screenshot 2024-09-19 at 4.13.31 AM.png

```python
model = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = model.detectMultiScale(img, 1.3, 7)
```

```python
faces.shape
```

```python
for i in faces:
    x, y, w, h = i
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 0), 10)

plt.imshow(img)
plt.show()
```
